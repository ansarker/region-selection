```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import csv
import os
```

```python
def select_regions(img):
    img_ = np.zeros(img.shape, dtype=np.uint8)
    img_[:, :, 0] = img[:, :, 2]
    img_[:, :, 1] = img[:, :, 1]
    img_[:, :, 2] = img[:, :, 0]
    
    fromCenter = False
    cv2.namedWindow('Select ROIs', cv2.WINDOW_NORMAL)
    roi = cv2.selectROIs('Select ROIs', img_, fromCenter)

    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return roi
```

```python
with open('selection-coordinates.csv', 'w') as csv_file:
    csv_writer = csv.writer(csv_file, delimiter=',')
    csv_writer.writerow(['Image', 'X', 'Y', 'W', 'H'])
```

```python
current_dir = os.getcwd()
classes = [x for x in os.listdir(os.path.join(current_dir, 'data'))]

# img_list = []
# img_path = []

# for cl in range(len(classes)):
#     for f in os.listdir(os.path.join('data', classes[cl])):
#         img_list.append(f)
#         img_path.append(os.path.join(os.path.join('data', classes[cl]), f))
```

```python
for i in range(len(classes)):
    if os.path.exists('Output/' + classes[i]):
        pass
    else:
        os.makedirs('Output/' + classes[i])
```

# All at once

```python
class_ind = 0
print(classes[class_ind])

for ind in os.listdir(os.path.join('data', classes[class_ind])):
    img_path = os.path.join('data', classes[class_ind]) + '/' + ind
    img = Image.open(img_path)
    img = np.array(img)
    roi = select_regions(img)
    
    num_of_selection = roi.shape[0]
    selection_list = []
    name =ind.split('.')

    with open('selection-coordinates.csv', 'a') as csv_file:
        csv_writer = csv.writer(csv_file, delimiter=',')
        for i in range(num_of_selection):
            img_name = name[0] + '_' + str(i) + '.' + name[1]
            X, Y, W, H = roi[i]
            selection_list.append(img[Y:Y+H, X:X+W])
            csv_writer.writerow([img_name, X, Y, W, H])
    
    bounding_box = img.copy()
    white_image = np.ones(bounding_box.shape)

    for i in range(num_of_selection):
        X, Y, W, H = roi[i]
        cv2.rectangle(white_image, (X, Y), (X+W, Y+H), (0, 0, 0), -1)
#         plt.figure(figsize=(12,12))
#         plt.imshow(bounding_box)
#         plt.show()
    cv2.imwrite(os.path.join('Output/' + classes[class_ind], ind), white_image*255)
    
```
